```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
 csv_file = "data-csv/penalty_kicks.csv"
penalty_kicks_df = pd.read_csv(csv_file)
penalty_kicks_df.head()
```

```python
new_penalty_kicks_df = penalty_kicks_df[['match_id', 'match_name','match_date', 'stage_name','team_id','team_name','player_id','family_name','given_name','shirt_number','converted']].copy()

new_penalty_kicks_df.head()
```

```python
new_penalty_kicks_df = new_penalty_kicks_df.loc[(new_penalty_kicks_df['match_date'] > '1997-12-31')] 
```

```python
knock_out = ['final','semi-finals','quarter-finals']

new_penalty_kicks_df = new_penalty_kicks_df.loc[new_penalty_kicks_df['stage_name'].isin(knock_out)]
```

```python
new_penalty_kicks_df['year'] = pd.DatetimeIndex(new_penalty_kicks_df['match_date']).year
```

```python
new_penalty_kicks_df.groupby(['team_name','year','family_name']).sum()
```
