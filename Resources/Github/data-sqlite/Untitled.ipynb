```python
import sqlite3
import pandas as pd
# Create your connection.
cnx = sqlite3.connect('worldcup.db')
```

```python
# create a cursor
cursor = cnx.cursor()

# execute the query
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# retrieve the results
results = cursor.fetchall()

# print the table names
for table in results:
    print(table[0])


```
