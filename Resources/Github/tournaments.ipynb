```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
 csv_file = "data-csv/tournaments.csv"
tournaments = pd.read_csv(csv_file)
tournaments.head()
```

```python
new_tournaments = tournaments[['tournament_id', 'tournament_name','year', 'host_country','winner','host_won','count_teams','group_stage','second_group_stage','final_round','round_of_16','quarter_finals','semi_finals','third_place_match','final']].copy()

new_tournaments.head()
```

```python
new_tournaments = new_tournaments.loc[(new_tournaments['year'] > 1997)]
new_tournaments.head()
```
